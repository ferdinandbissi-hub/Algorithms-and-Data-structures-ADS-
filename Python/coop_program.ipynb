### Exercise 2

Given an integer array nums, find the subarray with the largest sum, and return its sum

```python
import numpy as np
nums = np.array([1,2,3,-4,-6,9,10,4,5,7])
max_sum = 0
max_sub_arr = []
n = len(nums)
    
for i in range(n):
    max_sub_arr.append(nums[i])
    new_max_sum = max_sum + nums[i]
    if new_max_sum > max_sum:
        max_sum = new_max_sum
    else:
        max_sub_arr.remove(nums[i])
        continue
print(max_sub_arr, new_max_sum)  
```

### Exercice 3

 Given an array nums with n objects colored red, white, or blue, sort them inplace so that objects of the same color are adjacent, with the colors in the order red, white and blue

```python
import numpy as np
nums = np.array([0,1,1,2,2,2,2,0,0,1,2,2,1,1,0,0])

# Let r,w, and b related to colours red, white and blue
def sort_colored_object(nums)
    n = len(nums)
    r,w,b = 0,0,0

    for i in nums:
        if i != 0 and i != 1 and i != 2:
            print("Your array should contain only 0,1 a" \
            "nd 2 as objects")
            break
    else:
        # Count the number of red, white and blue objects
        for i in range(n):
            if nums[i] == 0:
                r = r+1
            elif nums[i] == 1:
                w = w+1
            elif nums[i] == 2:
                b = b+1
            else:
                continue

        idx = 0
        for i in range(r):
            nums[idx] = 0
            idx = idx+1
        for i in range(w):
            nums[idx] = 1
            idx = idx+1
        for i in range(b):
            nums[idx] = 2
            idx = idx+1
        print(nums)
    
```

Exercise 4: Given an array nums of n integers, return an array of all the unique quadruplets [nums[a], nums[b], nums[c], nums[d]] such that:
0<=a, b,c,d<n
a,b,c,and d are distinct
nums[a]+nums[b]+nums[c]+nums[d] == target

```python
import numpy as np

nums = np.array([2,4,5,1,0,2,5,9,10,13,15,22,7,8,29,45,3])
target = 100
n = len(nums)
for a in range(n):
    for b in range(a+1, n):
        for c in range(b+1, n):
            for d in range(c+1, n):
                if nums[a]+nums[b]+nums[c]+nums[d] == target:
                    print([nums[a],nums[b],nums[c],nums[d]])
                else:
                    continue
```

Exercise 5: Given an array of intervals where intervals[i] = [stati, endi], merge all overlapping intervals, and return and array of the non-overlapping intervals that cover all the intervals in the input.

```python
import numpy as np
inter = np.array([[1,3], [10,17], [4,9],[3,8]])
inter = inter[np.argsort(inter[:,0])]  # Sort based on the start point of each interval

print(inter)

result = []
for s,e in inter:
    if result and s <= result[-1][1]: # If result mean, does result has anything??
        result[-1][1] = max(result[-1][1], e)  # merge
    else:
        result.append([s,e])  # add

print(result)

```

Exercise 6: Given a string s of '(',')' and lowercase English characters.
Your task is to remove the minimum number of parentheses ('('or')'in any positions) so that the resulting parentheses string is valid and return any valid sting.

```python
import numpy as np

s = "))))(((("
result1 = []
count = 0

for i in s:
    if i == ")" and count == 0:
        continue

    # Count extra opening parenteses
    if i == "(":
        count += 1
    elif i == ")":
        count -= 1
    result1.append(i)
result = []
count1 = 0


for i in reversed(result1):
    if i == "(" and count1 == 0:
        continue

    if i == ")":
        count1 += 1
    elif i == "(":
        count1 -= 1
    result.append(i)
res = ''.join(reversed(result))
print(res)
    

```

Exercise 7: Given a string s, sort it in decreasing order based on the frequency of the characters. The frequency of a character is the number of times it appears in the string.
Return the sorted string. If there are multiple answers, return any of them

```python
import numpy as np
s = "caabbacbd"

dict = {} # store each character as a key and its frequency as the corresponding value

for i in s:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1
print(dict)

L = []
for i in dict:
    L.append((dict[i],i)) 
print(L)
L.sort(key=lambda x: x[0], reverse=True)
print(L)


result = ""
for i in range(len(L)):
    result = result + L[i][1]*L[i][0]
print(result)
```

Exercise 8: Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise. In the other words, return true if one of s1's permutations in the substring of s2

```python
import itertools    # Python's Itertool is a module that provides various functions that work on iterators to produce complex iterators

def truefalse(s1, s2):
    perm = itertools.permutations(s1)
    result = [''.join(p) for p in perm] #str.join() is used to combine elements of an iterable (like a list, tuple, or set) into a single string, using a specified separator between each element
    print(result)
    result.remove(result[0])
    print(result)

    for i in result:
        if i in s2:
            answer = True
        else:
            answer = False
    return answer
res = truefalse("abe","eidcbaooo")
print(res)


```
